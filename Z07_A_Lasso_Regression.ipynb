# Lasso Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn import linear_model
import scipy.stats as stats
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import model_selection
```

Rozważmy następujący zbiór punktów:

```python
true_fun = lambda X: np.cos(1.5 * np.pi * X)
n_samples=20
x = np.sort(np.random.rand(n_samples))
y = true_fun(x) + np.random.randn(n_samples) * 0.1
x=np.vstack(x)
```

# Przykład
Proszę wykonać regresję (Ridge Regression) na powyższym zbiorze danych 

```python
clf = Lasso(alpha=1.0)
clf.fit(x, y) 

x_plot = np.vstack(np.linspace(0, 1, 20))
plt.plot(x_plot, clf.predict(x_plot), color='blue',linewidth=3)
plt.plot(x, y, 'ok');
plt.show()
```

## Regresja liniowa

```python
# prepare models
models = []
predicts = []
names = []
models.append(('LR degree 2', make_pipeline(
    PolynomialFeatures(2), linear_model.LinearRegression())))
models.append(('LR degree 20', make_pipeline(
    PolynomialFeatures(20), linear_model.LinearRegression())))

x_plot = np.vstack(np.linspace(-3, 3, 1000))
for name, model in models:
    print(name)
    model.fit(x, y)
    predicts.append(model.predict(x_plot))
    names.append(name)

plt.plot(x, y, 'ok')
for i in range(len(models)):
    # print(i)
    plt.plot(x_plot, predicts[i], linewidth=3, label=names[i])
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
plt.legend()
plt.show()
```

# Zadanie 
Proszę wykonać 
* regresję dla wielomianów o stopniu **20**
* regresję (Lasso Regression) dla wielomianów o stopniu **20** oraz
   * alpha = 1 
   * alpha = 10 000 
   * alpha = 0.0001

```python
# prepare models
models = []
predicts = []
names = []
models.append(('LR degree 20', make_pipeline(
    PolynomialFeatures(20), linear_model.LinearRegression())))
models.append(('Ls alpha 1.0', make_pipeline(
    PolynomialFeatures(20), linear_model.Lasso(alpha=1))))
models.append(('Ls alpha 10000', make_pipeline(
    PolynomialFeatures(20), linear_model.Lasso(alpha=10000))))
models.append(('Ls alpha 0.0001', make_pipeline(
    PolynomialFeatures(20), linear_model.Lasso(alpha=0.0001))))

x_plot = np.vstack(np.linspace(-3, 3, 1000))
for name, model in models:
    print(name)
    model.fit(x, y)
    predicts.append(model.predict(x_plot))
    names.append(name)

plt.plot(x, y, 'ok')
for i in range(len(models)):
    # print(i)
    plt.plot(x_plot, predicts[i], linewidth=3, label=names[i])
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
plt.legend()
plt.show()
```

# Zdanie 
Dobierz optymalny stopień wielomianu oraz parametr alpha.

```python
seed=123
kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)
scoring = 'neg_mean_absolute_error'
#scoring = 'r2'
```

```python
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), linear_model.Lasso(alpha=1.0)),
                    param_grid={'polynomialfeatures__degree': [2, 3, 4, 5, 6, 25, 100],
                        'lasso__alpha': [10000, 1000, 100, 10, 1, 0.1, 0.01, 0.001, 0.0001, 0.00001, 0.000001]},
                    cv=kfold,
                    refit=True)
grid.fit(x, y)
grid.best_params_
print(grid.best_params_)
print("R2: {}".format(metrics.r2_score(y, grid.predict(x)) ))
```

# Zdanie 
Dobierz optymalny stopień wielomianu oraz parametr <tt>alpha</tt> za pomocą metody <tt>GridSearchCV</tt> dla danych reklamowych, obejmującym sprzedaż produktów i ich budżet reklamowy w trzech różnych mediach telewizyjnych, radiu, gazetach.

*  policz r_square score.

```python
import pandas as pd
df_adv = pd.read_csv('./Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
seed=123
kfold = model_selection.KFold(n_splits=10, random_state=seed, shuffle=True)
scoring = 'neg_mean_absolute_error'
#scoring = 'r2'
```

```python
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), linear_model.Lasso(alpha=1.0)),
                    param_grid={'polynomialfeatures__degree': [2,3,4],
                        'lasso__alpha': [0.00000001,0.000001,0.0001,1]},
                    cv=kfold,
                    scoring=scoring,
                    refit=True)
grid.fit(X, y)
grid.best_params_
print(grid.best_params_)
print("R2: {}".format(metrics.r2_score(y, grid.predict(X)) ))
```

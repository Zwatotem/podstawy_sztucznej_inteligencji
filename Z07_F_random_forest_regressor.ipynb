# Random Forest Regressor

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

Proszę narysować wykres dowolnej funkcji na odcinku $[0,3]$.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
x = stats.uniform(0,3).rvs(100)
y = f(x) + stats.norm(0,0.2).rvs(len(x))
```

# Zadanie
wykonaj regresję z różnymi parametrami

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 1000))
reg_rf1 = RandomForestRegressor(max_depth=5, n_estimators =5 ,random_state=2)
reg_rf2 = RandomForestRegressor(max_depth=10, n_estimators =5 ,random_state=2)
reg_rf3 = RandomForestRegressor(max_depth=1, n_estimators =5 ,random_state=2)
reg_rf4 = RandomForestRegressor(max_depth=100, n_estimators =5 ,random_state=2)
y_rbf1 = reg_rf1.fit(x, y)
y_rbf2 = reg_rf2.fit(x, y)
y_rbf3 = reg_rf3.fit(x, y)
y_rbf4 = reg_rf4.fit(x, y)
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, reg_rf1.predict(x_plot), color='blue',linewidth=3)
plt.plot(x_plot, reg_rf2.predict(x_plot), color='gray',linewidth=3)
plt.plot(x_plot, reg_rf3.predict(x_plot), color='green',linewidth=3)
plt.plot(x_plot, reg_rf4.predict(x_plot), color='black',linewidth=3)
plt.show()
```

# Zad

Wyznacz optymalne parametry dla modelu RandomForestRegressor.

```python
import pandas as pd
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
from sklearn import metrics
rf = GridSearchCV(RandomForestRegressor(max_depth=10, max_features=1, n_estimators=10, random_state=3), cv=5,
                   param_grid={'max_depth':    [1, 2, 3, 4, 5],
                               'max_features': [1, 2, 3, 4, 5],
                               'n_estimators': [1, 2, 3, 4, 5],
                               'random_state': [1, 2, 3, 4, 5]})
rf.fit(X, y)
print(rf.best_params_)
print("R2: {}".format(metrics.r2_score(y, rf.predict(X)) ))
```

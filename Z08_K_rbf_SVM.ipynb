```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

# RBF SVM Classification

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=.1, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

# Kernel trick

Naucz SVM z jÄ…drem "rbf" o parametrach

* gamma=0.01, C=1.0
* gamma=0.1, C=1.0

```python
svm = SVC(kernel='rbf', C=1, gamma=0.01)
svm.fit(X, y)
plot_decision_regions(X, y, clf=svm)
```

```python
svm = SVC(kernel='rbf', C=1, gamma=0.1)
svm.fit(X, y)
plot_decision_regions(X, y, clf=svm)
```

```python
svm = SVC(kernel='rbf', C=1, gamma=1)
svm.fit(X, y)
plot_decision_regions(X, y, clf=svm)
```

```python
svm = SVC(kernel='rbf', C=1, gamma=10)
svm.fit(X, y)
plot_decision_regions(X, y, clf=svm)
```

```python
svm = SVC(kernel='rbf', C=1, gamma=100)
svm.fit(X, y)
plot_decision_regions(X, y, clf=svm)
```
